```python
# 读取样本中的正负样本
import pandas as pd
```

```python
DF= pd.read_csv('training_data.tsv', sep='\t')
counts = DF['y'].value_counts()
counts,len(DF)
```

```python
df = pd.read_csv('all_data/met_r_training.tsv', sep='\t')
df['y'].value_counts()
```

```python
df_1 = df[df['y'] == 1].copy()
df_sample_1 = df_1.sample(frac=0.4, random_state=42)
# 负样本
df_0 = df[df['y'] == 0].copy()
df_sample_0 = df_0.sample(frac=0.4, random_state=42)

# 使用 frac 参数进行比例抽样
df_sample = pd.concat([df_sample_0, df_sample_1], axis=0, ignore_index=True)
```
